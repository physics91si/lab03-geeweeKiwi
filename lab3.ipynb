```python
# Don't edit this function
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample
```

```python
# Edit this function. "pass" tells Python to do nothing.
import string

def parse_sample(sample, words_only = False, sort_list = False):
    list =[]
    first_letter = 0
    sample = sample.replace('-',' ')
    sample = sample.replace('\n',' ')
    if words_only:
        sample = sample.translate(None, string.punctuation)
    for i in range(0,len(sample)):
        if sample[i] == ' ':
            str = sample[first_letter:i]
            list.append(str)
            if i < len(sample)-1 and sample[i+1] != ' ':
                first_letter = i+1
        if i == len(sample)-1 and not sample[i] == ' ':
            list.append(sample[first_letter:])
    if sort_list:
        list.sort()
    return list

parse_sample(load_sample(), True, True)
```

```python
import numpy as np

def count_freq(lst):
    """Return a dict of word frequencies given a list of words.
    
    Arguments:
    -- lst: a list of strings
    
    Returns:
    -- freq_dict: a dictionary of word: frequency pairs. The keys are strings
       containing the words, and the values are integers indicating how many
       times each word appears.
    
    Example:
    >>> count_freq(['time', 'after', 'time'])
    {'time': 2, 'after': 1}
    """
    wdict = {}
    for word in lst:
        if word in wdict:
            wdict[word] += 1
        else:
            wdict[word] = 1
    return wdict

def mean(lst):
    """Return the arithmetic mean of a list of numbers."""
    sum = 0.
    for num in lst:
        sum += num
    return sum/len(lst)
        

def stdev(lst):
    """Return the standard deviation of a list of numbers."""
    m = mean(lst)
    ssum = 0.
    for num in lst:
        ssum += (num - m)**2
    var = ssum / len(lst)
    return np.sqrt(var)

def median(lst):
    """Return the median of a list of numbers."""
    lst = sorted(lst)
    if len(lst)%2 == 1:
        return lst[len(lst)/2]
    else:
        return (lst[len(lst)/2]+lst[len(lst)/2 - 1])/2.

mean([2,6,19,65,4,3])
stdev([2,5,7,3])
median([1,2,3,4,5,6,20,-69.45])
```

```python
def print_stats(lst):
    """Print common statistics about the word frequencies of a list of words."""
    wdict = count_freq(lst)
    nlst = []
    for word in wdict:
        nlst.append(wdict[word])
    print('The Mean number of times a word appears is: '  + str(mean(nlst)))
    print('The Standard Deviation of the number of times a word appears is ' + str(stdev(nlst)))
    print('The Median number of times a word appears is: ' + str(median(nlst)))
print_stats(parse_sample(load_sample(), True, True))        
```
